# WeddingWire Messenger

This class provides automated tools for interacting with WeddingWire's messaging system. It allows you to search through chat histories with venues and send messages to venues directly. The class uses browser automation to handle the interactions, making it possible to programmatically manage your WeddingWire communications. Note that you'll need to be logged into WeddingWire (via cookies) to use these features.

```python
messenger = WeddingWireMessenger(ui=False)
```

# Task #1 - Send a custom message to a venue

```python
VENUE_NAME = "Riverside Farm"
MSG = "We're still very interested in your venue for our wedding on Feb 1, 2027. Could you please send over more details about available packages and what's included?"

task_send_message = f"""
Go to weddingwire.com/users-login.php and make sure that we are logged in.
Wait for the dashboard to load.
Go to Inbox (messages).
Open the conversation with the venue named {VENUE_NAME} in the inbox.
Send the following message:
'{MSG}'
Close the browser.
"""
```

```python
task_send_message
```

```python
async def main():
    # Configure the browser to connect to your Chrome instance
    browser = Browser(
        config=BrowserConfig(
            browser_binary_path='/Applications/Google Chrome.app/Contents/MacOS/Google Chrome'  
        )
    )

    # Create the agent with the configured browser
    agent = Agent(
        task=task_send_message,  # Your task variable
        llm=ChatOpenAI(model="gpt-4o"),
        browser=browser
    )

    # Run the agent
    await agent.run()

    # Optional: Keep the browser open until user input
    input('Press Enter to close the browser...')
    await browser.close()

await main()
```
